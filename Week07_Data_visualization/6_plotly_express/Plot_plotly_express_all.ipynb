## Introduction to plotly.express

**plotly.express** (often imported as px) is a high-level plotting library built on top of Plotly.
It allows you to create interactive, beautiful charts in just one line of code.

**Key Features**

* Interactive charts by default

* Supports multiple chart types: scatter, line, bar, pie, histogram, box, violin, bubble, maps, and more

* Works directly with Pandas DataFrames  no complex setup

* Customizable: colors, labels, titles, hover info

import plotly.io as pio

pio.renderers.default = "svg"

| Renderer Name | Used For             | Description                                               |
| ------------- | -------------------- | --------------------------------------------------------- |
| `"notebook"`  | Jupyter Notebook     | Default for inline charts in Jupyter                      |
| `"browser"`   | Opens in browser tab | Renders chart in your web browser                         |
| `"svg"`       | Static vector image  | Displays chart as **static SVG image** (no interactivity) |
| `"png"`       | Static raster image  | Shows a static PNG image                                  |
| `"iframe"`    | Embedded HTML        | Embeds interactive chart inside a frame                   |
| `"colab"`     | Google Colab         | For Colab environment                                     |


```python
import plotly.express as px
import pandas as pd
import numpy as np

import plotly.io as pio
pio.renderers.default = "svg" #Static vector image	Displays chart as static SVG image (no interactivity)
#pio.renderers.default = "notebook"  #Default for inline charts in Jupyter
```

# 1 Scatter Plot

```python
# Sample dataset
df_scatter = pd.DataFrame({
    'Height': [150, 160, 170, 180, 190],
    'Weight': [50, 60, 65, 80, 90],
    'Gender': ['F', 'F', 'M', 'M', 'M']
})

fig = px.scatter(df_scatter, x='Height', y='Weight', color='Gender',
                 title='Scatter Plot: Height vs Weight')
fig.show()
```

# 2 Line Chart

```python
df_line = pd.DataFrame({
    'Month': ['Jan','Feb','Mar','Apr','May'],
    'Sales': [200, 250, 300, 280, 350]
})

fig = px.line(df_line, x='Month', y='Sales', title='Monthly Sales Trend')
fig.show()
```

# 3 Bar Chart

```python
df_bar = pd.DataFrame({
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango'],
    'Quantity': [10, 15, 7, 20]
})

fig = px.bar(df_bar, x='Fruit', y='Quantity', color='Fruit', title='Fruit Quantity')
fig.show()
```

# 4 Pie Chart

```python
df_bar = pd.DataFrame({
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango'],
    'Quantity': [10, 15, 7, 20]
})
fig = px.pie(df_bar, names='Fruit', values='Quantity', title='Fruit Distribution')
fig.show()
```

# 5 Donut Chart

```python
df_bar = pd.DataFrame({
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango'],
    'Quantity': [10, 15, 7, 20]
})
fig = px.pie(df_bar, names='Fruit', values='Quantity', hole=0.4, title='Donut Chart Example')
fig.show()
```

# 6 Histogram

```python
df_hist = pd.DataFrame({'Scores': [45, 55, 65, 75, 85, 95, 55, 65, 75, 85]})
fig = px.histogram(df_hist, x='Scores', nbins=5, title='Score Distribution')
fig.show()
```

# 7 Box Plot

```python

# Simple dataset
df_simple = pd.DataFrame({
    'Class': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Score': [75, 85, 80, 90, 60, 80, 60, 70, 85]
})
fig = px.box(df_simple, x='Class', y='Score', title='Box Plot of Scores')
fig.show()
```

# 8 Violin Plot

```python
# Simple dataset
df_simple = pd.DataFrame({
    'Class': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Score': [75, 85, 80, 90, 60, 80, 60, 70, 85]
})
fig = px.violin(df_simple, x='Class', y='Score', box=True, points='all', title='Violin Plot of Scores')
fig.show()
```

# 9 Bubble Chart

```python
df_bubble = pd.DataFrame({
    'Country': ['A', 'B', 'C', 'D'],
    'GDP': [1000, 2000, 3000, 4000],
    'LifeExp': [60, 70, 65, 75],
    'Population': [1.5, 2.5, 3.5, 4.5]
})

fig = px.scatter(df_bubble, x='GDP', y='LifeExp', size='Population', color='Country',
                 hover_name='Country', title='Bubble Chart: GDP vs Life Expectancy')
fig.show()
```

#  Treemap

```python
df_tree = pd.DataFrame({
    'Continent': ['Asia','Asia','Europe','Europe'],
    'Country': ['India','China','France','Germany'],
    'Population': [1400, 1300, 67, 83]
})

fig = px.treemap(df_tree, path=['Continent','Country'], values='Population',
                 title='Treemap of Population by Continent')
fig.show()
```

# 11 Sunburst Chart

```python
df_tree = pd.DataFrame({
    'Continent': ['Asia','Asia','Europe','Europe'],
    'Country': ['India','China','France','Germany'],
    'Population': [1400, 1300, 67, 83]
})
fig = px.sunburst(df_tree, path=['Continent','Country'], values='Population',
                  title='Sunburst Chart of Population')
fig.show()
```

# 12 Area Chart

```python
df_area = pd.DataFrame({
    'Month': ['Jan','Feb','Mar','Apr'],
    'ProductA': [10, 15, 20, 25],
    'ProductB': [5, 7, 10, 12]
})
fig = px.area(df_area, x='Month', y=['ProductA','ProductB'], title='Area Chart Example')
fig.show()
```

# 13 Heatmap

```python
data = np.random.rand(5,5)
fig = px.imshow(data, text_auto=True, title='Random Heatmap')
fig.show()
```

# 14 Density Contour

```python
df_density = pd.DataFrame({
    'X': np.random.randn(100),
    'Y': np.random.randn(100)
})
fig = px.density_contour(df_density, x='X', y='Y', title='Density Contour Plot')
fig.show()
```

# 15 Map

```python
df_map = pd.DataFrame({
    'Country': ['USA','CAN','MEX'],
    'ISO': ['USA','CAN','MEX'],
    'Population': [300, 37, 120]
})

fig = px.scatter_geo(df_map, locations='ISO', color='Country', size='Population',
                     projection='natural earth', title='Sample Map')
fig.show()
```

# 16 Animated Scatter

```python
df_anim = pd.DataFrame({
    'Year': [2000,2000,2001,2001],
    'Country': ['A','B','A','B'],
    'GDP': [100,200,150,250],
    'LifeExp':[60,70,62,72]
})

fig = px.scatter(df_anim, x='GDP', y='LifeExp', animation_frame='Year', animation_group='Country',
                 size='GDP', color='Country', hover_name='Country', log_x=False, size_max=50,
                 title='Animated Scatter Example')
fig.show()
```
