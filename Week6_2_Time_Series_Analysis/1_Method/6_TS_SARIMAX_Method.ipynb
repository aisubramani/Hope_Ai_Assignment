‚≠ê 6Ô∏è‚É£ SARIMAX (SARIMA + Exogenous Variables)

Where X‚Çú = extra inputs, like:
‚úî Volume
‚úî High/Low
‚úî Temperature
‚úî Holidays
‚úî Other market indicators

üìå Example:
Close price depends on:
Open
High
Low
Volume
‚Üí SARIMAX learns this better than SARIMA.

When to use:
Trend + Seasonality + External features influence the results.

```python
# ========================================================
# 0Ô∏è‚É£ IMPORT LIBRARIES
# ========================================================
import numpy as np
import yfinance as yf
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt


# ========================================================
# 1Ô∏è‚É£ DOWNLOAD DATA
# ========================================================
ticker = yf.Ticker("TATACONSUM.NS")
stk_data = ticker.history(start="2024-01-01", end="2025-01-03")

print(stk_data)
stk_data.info()


# ========================================================
# 2Ô∏è‚É£ PREPARE TARGET & EXOGENOUS VARIABLES
# ========================================================
y = stk_data["Close"].values.reshape(-1, 1)
X = stk_data[["Open", "High", "Low", "Volume"]].values  # Exogenous variables

# Normalize BOTH y and X
scaler_y = MinMaxScaler()
scaler_x = MinMaxScaler()

data_y = scaler_y.fit_transform(y).flatten()
data_x = scaler_x.fit_transform(X)

print("Shapes -> y:", data_y.shape, " X:", data_x.shape)


# ========================================================
# RMSE + MAPE FUNCTION
# ========================================================
def rmsemape(actual, predicted):
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = mean_absolute_percentage_error(actual, predicted)
    return rmse, mape


# ========================================================
# 3Ô∏è‚É£ DEFINE SARIMAX ORDERS TO TEST
# ========================================================
orders = [(1,1,1), (1,1,2), (2,1,1)]
seasonal_orders = [(1,1,1,7), (1,1,2,7)]

results = []

print("\n==============================")
print("Testing SARIMAX Models...")
print("==============================\n")


# ========================================================
# 4Ô∏è‚É£ TRAIN SARIMAX MODELS
# ========================================================
for order in orders:
    for seas in seasonal_orders:

        print(f"Testing SARIMAX{order}x{seas} with exogenous features...")

        try:
            model = SARIMAX(
                data_y,
                exog=data_x,
                order=order,
                seasonal_order=seas,
                enforce_stationarity=False,
                enforce_invertibility=False
            )

            model_fit = model.fit(disp=False)

            # Predict full training data
            y_pred = model_fit.predict(0, len(data_y)-1, exog=data_x)

            rmse, mape = rmsemape(data_y, y_pred)

            print(f"Order={order}, Seasonal={seas}, RMSE={rmse:.6f}, MAPE={mape:.6f}")
            print("---------------------------------------")

            results.append({
                'order': order,
                'seasonal': seas,
                'rmse': rmse,
                'mape': mape
            })

        except Exception as e:
            print(f"Failed for SARIMAX{order}x{seas} : {e}")
            print("---------------------------------------")


# ========================================================
# 5Ô∏è‚É£ SELECT BEST SARIMAX MODEL
# ========================================================
best_model = min(results, key=lambda x: x["rmse"])

best_order = best_model["order"]
best_seasonal = best_model["seasonal"]

print("\n==============================")
print("üî• BEST SARIMAX MODEL SELECTED")
print("==============================")
print(f"Order          : {best_order}")
print(f"Seasonal Order : {best_seasonal}")
print(f"RMSE           : {best_model['rmse']:.6f}")
print(f"MAPE           : {best_model['mape']:.6f}")


# ========================================================
# 6Ô∏è‚É£ FIT BEST SARIMAX MODEL (FULL DATA)
# ========================================================
model = SARIMAX(
    data_y, 
    exog=data_x,
    order=best_order,
    seasonal_order=best_seasonal,
    enforce_stationarity=False,
    enforce_invertibility=False
)

model_fit = model.fit()
fitted = model_fit.predict(0, len(data_y)-1, exog=data_x)


# ========================================================
# 7Ô∏è‚É£ PLOT ACTUAL vs FITTED
# ========================================================
plt.figure(figsize=(14,6))
plt.plot(data_y, label="Actual (Scaled)", color="blue")
plt.plot(fitted, label=f"Fitted SARIMAX{best_order}x{best_seasonal}", color="green")

plt.title("TATACONSUM.NS - SARIMAX Training Fit (with Exogenous Features)")
plt.xlabel("Days")
plt.ylabel("Scaled Close Price")
plt.legend()
plt.grid()
plt.show()


# ========================================================
# 8Ô∏è‚É£ 30-DAY FORECAST (with exogenous future values)
# ========================================================

future_steps = 30

# Use last known exogenous value repeated (simple method)
last_exog = data_x[-1].reshape(1, -1)
future_exog = np.repeat(last_exog, future_steps, axis=0)

scaled_forecast = model_fit.predict(
    start=len(data_y),
    end=len(data_y)+future_steps-1,
    exog=future_exog
)

# Convert to real price
scaled_forecast = scaled_forecast.reshape(-1,1)
real_forecast = scaler_y.inverse_transform(scaled_forecast)


print("\n==============================")
print("üìà NEXT 30 DAYS SARIMAX FORECAST (REAL PRICES)")
print("==============================\n")

for i, val in enumerate(real_forecast):
    print(f"Day +{i+1} : ‚Çπ{val[0]:.2f}")


# ========================================================
# 9Ô∏è‚É£ FINAL FORECAST PLOT
# ========================================================
plt.figure(figsize=(14,6))

plt.plot(data_y, label="Actual", color="blue")
plt.plot(fitted, label="Fitted", color="green")

forecast_index = range(len(data_y), len(data_y)+future_steps)
plt.plot(forecast_index, scaled_forecast, label="30-Day Forecast (Scaled)", color="red")

plt.title("TATACONSUM.NS - SARIMAX Forecast (30 Days)")
plt.xlabel("Days")
plt.ylabel("Scaled Close Price")
plt.legend()
plt.grid()
plt.show()
```
