‚≠ê 5Ô∏è‚É£ SARIMA (Seasonal ARIMA)

Adds seasonality.

Where:

s = season length (7 for weekly, 12 for monthly)

(P,D,Q) = seasonal AR, differencing, MA

üìå Example:
Sales rise every weekend ‚Üí weekly seasonality (s=7)

When to use:
Data has trend + seasonality.

```python
# ========================================================
# 0Ô∏è‚É£ IMPORT LIBRARIES
# ========================================================
import numpy as np
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt


# ========================================================
# 1Ô∏è‚É£ DOWNLOAD DATA
# ========================================================
ticker = yf.Ticker("TATACONSUM.NS")
stk_data = ticker.history(start="2024-01-01", end="2025-01-03")

print(stk_data)
stk_data.info()


# ========================================================
# 2Ô∏è‚É£ NORMALIZE CLOSE PRICE
# ========================================================
column = "Close"
Ms = MinMaxScaler()

data1 = Ms.fit_transform(stk_data[[column]])
data1 = data1.flatten()  # SARIMA needs 1D array


# ========================================================
# RMSE + MAPE FUNCTION
# ========================================================
def rmsemape(actual, predicted):
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = mean_absolute_percentage_error(actual, predicted)
    return rmse, mape


# ========================================================
# 3Ô∏è‚É£ DEFINE SARIMA ORDERS
# ========================================================
orders = [(1,1,1), (1,1,2), (2,1,1), (2,1,2)]
seasonal_orders = [(1,1,1,7), (1,1,2,7), (2,1,1,7), (2,1,2,7)]

results = []

print("\n==========================")
print("Testing SARIMA Models...")
print("==========================\n")


# ========================================================
# 4Ô∏è‚É£ LOOP THROUGH SARIMA MODELS
# ========================================================
for order in orders:
    for seas in seasonal_orders:

        print(f"Testing SARIMA{order}x{seas} ...")

        try:
            model = SARIMAX(
                data1,
                order=order,
                seasonal_order=seas,
                enforce_stationarity=False,
                enforce_invertibility=False
            )
            model_fit = model.fit(disp=False)

            # Predict full dataset
            y_pred = model_fit.predict(0, len(data1)-1)

            # Compute performance
            rmse, mape = rmsemape(data1, y_pred)

            print(f"Order={order}, Seasonal={seas}, RMSE={rmse:.6f}, MAPE={mape:.6f}")
            print("---------------------------------------")

            results.append({
                'order': order,
                'seasonal': seas,
                'rmse': rmse,
                'mape': mape
            })

        except Exception as e:
            print(f"Model failed for SARIMA{order}x{seas}: {e}")
            print("---------------------------------------")


# ========================================================
# 5Ô∏è‚É£ SELECT BEST SARIMA MODEL
# ========================================================
best_model = min(results, key=lambda x: x['rmse'])

best_order = best_model['order']
best_seasonal = best_model['seasonal']

print("\n==============================")
print("üî• BEST SARIMA MODEL SELECTED")
print("==============================")
print(f"Order          : {best_order}")
print(f"Seasonal Order : {best_seasonal}")
print(f"RMSE           : {best_model['rmse']:.6f}")
print(f"MAPE           : {best_model['mape']:.6f}")


# ========================================================
# 6Ô∏è‚É£ FIT BEST SARIMA MODEL AND PLOT TRAINING RESULTS
# ========================================================
model = SARIMAX(
    data1,
    order=best_order,
    seasonal_order=best_seasonal,
    enforce_stationarity=False,
    enforce_invertibility=False
)

model_fit = model.fit()

fitted = model_fit.predict(0, len(data1)-1)

plt.figure(figsize=(12,6))
plt.plot(data1, label="Actual", color="blue")
plt.plot(fitted, label=f"Fitted SARIMA{best_order}x{best_seasonal}", color="green")

plt.title("TATACONSUM.NS - SARIMA Fitted Model")
plt.xlabel("Days")
plt.ylabel("Scaled Price")
plt.legend()
plt.grid()
plt.show()


# ========================================================
# 7Ô∏è‚É£ FORECAST NEXT 30 DAYS
# ========================================================
future_steps = 30

future_scaled = model_fit.predict(
    start=len(data1),
    end=len(data1) + future_steps - 1
)

# Inverse transform to real price
future_scaled = future_scaled.reshape(-1,1)
future_real = Ms.inverse_transform(future_scaled)

print("\n==============================")
print("üìà NEXT 30 DAYS SARIMA FORECAST")
print("==============================\n")

for i, val in enumerate(future_real):
    print(f"Day +{i+1} : ‚Çπ{val[0]:.2f}")


# ========================================================
# 8Ô∏è‚É£ PLOT FULL FIT + FORECAST
# ========================================================
plt.figure(figsize=(14,6))

# Actual
plt.plot(data1, color="blue", label="Actual")

# Fitted
plt.plot(fitted, color="green", label="Fitted")

# Forecast (Scaled)
forecast_index = range(len(data1), len(data1) + future_steps)
plt.plot(forecast_index, future_scaled, color="red", label="30-Day Forecast")

plt.title("TATACONSUM.NS - SARIMA Forecasting (30 Days)")
plt.xlabel("Days")
plt.ylabel("Scaled Price")
plt.legend()
plt.grid()
plt.show()
```
