# Q1. ValueError in Rolling Mean Calculation

```python
import pandas as pd

df = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03'],
    'Sales': [100, 120, 130]
})

df['7_day_avg'] = df['Sales'].rolling(window=7, min_periods=1).mean()
print(df)
```

# Q2.KeyError for Time Series Column

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# -------------------------
# 1. Sample Data
# -------------------------
data = {
    'Date': ['2023-01-01','2023-01-02','2023-01-03','2023-01-04',
             '2023-01-05','2023-01-06','2023-01-07'],
    'Temp': [30, 32, 31, 29, 28, 27, 26]
}

df = pd.DataFrame(data)

# -------------------------
# 2. Convert Date and set index
# -------------------------
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

# -------------------------
# 3. ARIMA Model
# -------------------------
model = ARIMA(df['Temp'], order=(1, 1, 1))
model_fit = model.fit()

# -------------------------
# 4. Model Summary
# -------------------------
print(model_fit.summary())

# -------------------------
# 5. Forecast next 5 values
# -------------------------
forecast = model_fit.forecast(steps=5)
print("\nForecasted values:")
print(forecast)
```

# Q4. NameError

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Load dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url, header=0, parse_dates=['Month'], index_col='Month')

# Fit ARIMA model
model = ARIMA(df['Passengers'], order=(5, 1, 0))
model_fit = model.fit()   #  Correct

# Forecast next 12 time steps
forecast = model_fit.forecast(steps=12)

# Plot forecast
plt.plot(df['Passengers'], label="Original")
plt.plot(forecast, label="Forecast", linestyle='--')
plt.legend()
plt.show()
```
