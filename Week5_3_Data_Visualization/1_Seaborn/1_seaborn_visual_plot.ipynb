<h2 style="color:purple"> Univarite Analysis - Seaborn</h2>

```python
# Histogram, density plot -red
sb.distplot(dataset['ssc_p'], hist=True,kde=True, kde_kws={'color':"red"})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="Set2")
plt.figure(figsize=(8,5))
sns.histplot(dataset['ssc_p'], kde=True)
plt.title("Mark Distribution", fontsize=14, color="blue")
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt

# Catplot replaces factorplot
sb.catplot(
    x="gender",          # x-axis category
    y="ssc_p",           # numeric variable
    hue="status",        # separate by work experience
    kind="violin",       # violin plot
    col="degree_t",      # separate plots by degree type
    data=dataset
)

plt.show()
```

```python
#sns.regplot() is used to plot a scatterplot of two variables with a regression line.
import seaborn as sns
import matplotlib.pyplot as plt

# Regression plot: SSC vs HSC percentage
sns.regplot(x="ssc_p", y="hsc_p", data=dataset)
plt.title("SSC vs HSC Percentage with Regression Line")
plt.xlabel("SSC Percentage")
plt.ylabel("HSC Percentage")
plt.show()
```

```python
#lmplot = â€œlinear model plotâ€.
#Similar to regplot (scatter + regression line), but more powerful for faceting.
import seaborn as sns
import matplotlib.pyplot as plt

# Linear model plot: SSC vs HSC percentage, separate by gender
sns.lmplot(x="ssc_p", y="hsc_p", data=dataset, hue="gender", ci=None)
plt.title("SSC vs HSC Percentage by Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter dataset using query
female_data = dataset.query("gender == 'F'")

# Linear model plot for Female students only
sns.lmplot(x="ssc_p", y="hsc_p", data=female_data, ci=None )
plt.title("SSC vs HSC Percentage (Female Students)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter dataset using query
female_data = dataset.query("gender == 'F'")

# Linear model plot for Female students only
sns.lmplot(x="ssc_p", y="hsc_p", data=female_data, ci=None,order=2 )
plt.title("SSC vs HSC Percentage (Female Students)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create FacetGrid: one subplot per degree type
g = sns.FacetGrid(dataset, col="degree_t", hue="gender", height=4, aspect=1)
g.map(sns.scatterplot, "ssc_p", "hsc_p")
g.add_legend()
plt.show()
```

```python
#PairGrid lets you plot multiple relationships in a grid: every numeric variable against every other numeric variable
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns
cols = ["ssc_p", "hsc_p", "degree_p"]

# Create PairGrid
g = sns.PairGrid(dataset[cols], height=3)
g.map_upper(sns.scatterplot)   # upper triangle = scatter
g.map_lower(sns.kdeplot, cmap="Blues_d")  # lower triangle = density
g.map_diag(sns.histplot, bins=10, color="skyblue")  # diagonal = histogram

plt.show()
```
