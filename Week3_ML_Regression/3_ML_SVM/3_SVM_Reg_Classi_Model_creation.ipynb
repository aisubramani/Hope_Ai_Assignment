```python
import pandas as pd
dataset=pd.read_csv("50_Startups.csv")
```

```python
#data preprocessing
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset.columns
```

```python
#split input and output
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
```

```python
#split train and test
from sklearn.model_selection import train_test_split
#call 
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#standardistion
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html
#procoure
from sklearn.svm import SVR
#shape linear=strieght vbf=round #sigmoid=round #poly=U 
#kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} or callable, default=’rbf’
#regressor=SVR(kernel="sigmoid",C=100) 
regressor=SVR(kernel="linear",C=100000) 
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
#find how many support vector (data point) in this SVM
regressor.n_support_
```

```python
#find list of how many support vector (data point) in this SVM
regressor.support_
```

```python
# 5 Evaluation metrics
y_pred=regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename="finalized_model_SVM.sav"
pickle.dump(regressor,open(filename,'wb'))
loaded_model=pickle.load(open("finalized_model_SVM.sav",'rb'))
result=loaded_model.predict([[1234,345,4565,1,0]])
result
```
