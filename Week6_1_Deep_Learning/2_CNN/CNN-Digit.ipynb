1) What this script does (short)

You build and train a Convolutional Neural Network (CNN) to classify handwritten digits from the MNIST dataset (28×28 grayscale images). The CNN learns spatial patterns (strokes, edges) using convolutional filters, pools to reduce spatial size, then uses dense layers to classify into 10 classes (digits 0–9).

2) Quick glossary (one-liners)

Conv2D(filters, kernel_size): learns small local patterns (a filter/kernel slides over the image).

MaxPooling2D(pool_size): reduces spatial size (keeps important features, lowers computation).

Flatten(): turns 2D feature maps into a 1D vector for Dense layers.

Dense(units, activation): fully connected neural layer (classification happens here).

Dropout(rate): randomly disables neurons during training to reduce overfitting.

softmax: turns final outputs into probabilities for each class.

sparse_categorical_crossentropy: loss used when labels are integers (0..9).

3) Walkthrough of your code (line-by-line explanation)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

# 1) Load data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# 2) Reshape to add channel
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32')
x_test  = x_test.reshape(-1, 28, 28, 1).astype('float32')

# 3) Normalize (0..1)
x_train /= 255.0
x_test  /= 255.0

'''
You build and train a Convolutional Neural Network (CNN) to classify handwritten 
digits from the MNIST dataset (28×28 grayscale images). The CNN learns spatial 
patterns (strokes, edges) using convolutional filters, pools to reduce spatial size, 
then uses dense layers to classify into 10 classes (digits 0–9). 

Conv2D(filters, kernel_size): learns small local patterns (a filter/kernel slides over the image).
MaxPooling2D(pool_size): reduces spatial size (keeps important features, lowers computation).

Flatten(): turns 2D feature maps into a 1D vector for Dense layers.

Dense(units, activation): fully connected neural layer (classification happens here).

Dropout(rate): randomly disables neurons during training to reduce overfitting.

softmax: turns final outputs into probabilities for each class.

sparse_categorical_crossentropy: loss used when labels are integers (0..9).

'''

# 4) Build a stronger small CNN
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),  # detect simple edges
    BatchNormalization(),
    Conv2D(32, (3,3), activation='relu'),                         # combine into more complex features
    MaxPooling2D((2,2)),                                          # reduce spatial size
    Dropout(0.25),

    Conv2D(64, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.25),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')  # 10 classes
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# 5) Train with a validation split and callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
]

history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_split=0.1,
                    callbacks=callbacks)

# 6) Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test accuracy:", test_acc)

# 7) Predict & show probabilities for one image
import matplotlib.pyplot as plt
idx = 6900
plt.imshow(x_test[idx].reshape(28,28), cmap='Greys')
plt.axis('off')
probs = model.predict(x_test[idx:idx+1])[0]   # shape (10,)
print("Predicted class:", probs.argmax(), " Probabilities:", probs.round(3))
```

```python
import tensorflow as tf
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
input_shape = (28, 28, 1)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train /= 255
x_test /= 255

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
model = Sequential()
model.add(Conv2D(28, kernel_size=(3,3), input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation=tf.nn.relu))
model.add(Dropout(0.2))
model.add(Dense(10,activation=tf.nn.softmax))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x=x_train,y=y_train, epochs=10)

model.evaluate(x_test, y_test)

import matplotlib.pyplot as plt
image_index = 6900
plt.imshow(x_test[image_index].reshape(28, 28),cmap='Greys')
predict = x_test[image_index].reshape(28,28)
pred = model.predict(x_test[image_index].reshape(1, 28, 28, 1))
print(pred.argmax())
```
